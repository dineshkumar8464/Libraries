#Marker


*.   It can represent the values by different ways. They are

* STYLES:

1.  o  - Circle
2.  s  - Square
3.  D  - Diamond
4.  h  - Hexagon
5.  p  - Pentagon
6.  (*)  - Star
7.  |  - Vline
8.  _  - Hline
9.  ^  - Triangle

Colors : Represents in Hexadecimal or single character
1.  mec - marker_edge_color  - outside color of the style
2.  mfc - marker_face_color  - inside color of the style
3.  ms - marker_size         - size of the style




```python
x=np.array([1,2,3,4,5])
y=np.array([10,20,30,40,50])
plt.plot(x,y,label="Line chart",marker="o",color="red",mec='k',mfc ='w',ms='9')
plt.legend()
plt.title("Example of a line chart")
plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.show()
```

#fontdict

*  f1={"family" : "arial","color":"red", "size" : 19}

1.   title(" " ,fontdict=f1)
2.   xlabel(" ",fontdict=f1)
3.   ylabel(" ",fontdict=f1)



```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]
plt.plot(x,y,label="line chart")

#fontdict
f1={"family" : "italic","color":"red", "size" : 30}
f2={"family" : 'italic',"color":"green", "size" : 19}
plt.title("graph",fontdict=f1)
plt.xlabel("x -values",fontdict=f2)
plt.ylabel("y -values",fontdict=f2)
plt.legend()
```

legend with parameter

```python
#legend with parameter
x=np.arange(0,10,2)
y1=x**2
y2=x**3
plt.plot(x,y1,label="value1")
plt.plot(x,y2,label="value2")
plt.legend(loc='upper right',framealpha=0.9,edgecolor="red",facecolor="blue",shadow=True,fancybox=True)
#plt.legend(loc=0,framealpha=0.9,edgecolor="red",facecolor="blue",shadow=True,fancybox=True)
#plt.legend(loc=1,framealpha=0.9,edgecolor="red",facecolor="blue",shadow=True,fancybox=True)

#upto

#plt.legend(loc=10,framealpha=0.9,edgecolor="red",facecolor="blue",shadow=True,fancybox=True)

#and

#plt.legend(loc='upper left',framealpha=0.9,edgecolor="red",facecolor="blue",shadow=True,fancybox=True)
#plt.legend(loc='upper right',framealpha=0.9,edgecolor="red",facecolor="blue",shadow=True,fancybox=True)
#plt.legend(loc='lower left',framealpha=0.9,edgecolor="red",facecolor="blue",shadow=True,fancybox=True)
#plt.legend(loc='lower right',framealpha=0.9,edgecolor="red",facecolor="blue",shadow=True,fancybox=True)
#plt.legend(loc='center',framealpha=0.9,edgecolor="red",facecolor="blue",shadow=True,fancybox=True)
#plt.legend(loc='lower center',framealpha=0.9,edgecolor="red",facecolor="blue",shadow=True,fancybox=True)
plt.show()
```

#stacked bar plot

```python
#stacked bar plot
years =["2017","2018","2019","2020","2021"]
placements =["287","250","220","200","180"]

cse=[30,34,43,40,36]
ece=[35,38,45,45,35]
mech=[46,36,45,40,35]

plt.bar(years,placements,label="placements")
plt.bar(years,cse,label="cse")
plt.bar(years,ece,label="ece")
plt.bar(years,mech,label="mech")
plt.legend()
plt.show()
#
```

```python
#stacked bar plot
years =["2017","2018","2019","2020","2021"]
placements =["287","250","220","200","180"]

cse=[30,34,43,40,36]
ece=[35,38,45,45,35]
mech=[46,36,45,40,35]

plt.bar(years,placements,label="placements")
plt.bar(years,cse,label="cse")
plt.bar(years,ece,bottom =cse,label="ece")
plt.bar(years,mech,bottom=ece, label="mech")
plt.legend()
plt.show()
```

```python
#stacked bar plot
years =["2017","2018","2019","2020","2021"]
placements =["287","250","220","200","180"]

cse=[30,34,43,40,36]
ece=[35,38,45,45,35]
mech=[46,36,45,40,35]

ece_start=[cse[i]+ece[i] for i in range(len(cse))]

plt.title("years and placements")
plt.xlabel("years")
plt.ylabel("placements")


plt.bar(years,cse,label="cse",color='red')
plt.bar(years,ece,bottom =cse,label="ece",color="k")
plt.bar(years,mech,bottom=ece_start, label="mech",color='green')
plt.legend()
plt.show()
```
