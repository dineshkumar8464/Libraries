```python
import pandas as pd
s = pd.Series([1,2,3,4,5,6])
print(s)
```

```python
si = pd.Series([1,2,3,4,5,6],index =['a','b','c','d','e','f'])

print("index of series : ",si.index)                                         # index of series, returns only index values

print("values of series : ",s.values)                                         # values of series, returns only values without index

data = pd.Series([1,2,3,4,5,6],name ="number")

print("name of series :",data.name)                                         # name of series, the variable name must should be "name"

print("shape of series :",s.shape)                                          # shape of series, it is one dimesnional so it gives only 6

print("size of series :",s.size)                                            # no of elements in series

print("nbytes of series :",s.nbytes)                                        # memory occupied by series

print("dtype of series :",s.dtype)                                          # dataype of series

print("memory usage of values and index of series :/n ",s.memory_usage())   # memory usage of both values and index

print("memory usage of values of series : ",si.memory_usage(index=False))   #memory usage of values only

print("empty of series :",s.empty)

print("ndim of series :",s.ndim)           # no of dimesnions Series has only one dimension

s1 = pd.Series([])
print("the series is empty :", s1.empty)
```

# Removing Duplicates

2. data_frame.drop_duplicates( )
3. data_frame.drop_duplicates(inplace=True)





the dataset is not correct, so we can go with ur example for removing duplicates

```python
import pandas as pd
student_dict = {
    "name": ["Joe", "Nat", "Harry","dinesh","ramu",'ramu','sekhar','raja'],
    "roll_no": [101, 102, 103,109,108,104,102,100],
    "telugu": [85.10, 85.10, 91.54,76.4,88.9,89.4,57.6,48.6],
    "hindhi": [85.10, 77.80, 91.54,76.4,88.9,89.4,57.6,48.6],
    "english": [85.10, 77.80, 91.54,76.4,88.9,89.4,57.6,48.6],
    "maths": [85.10, 77.80, 91.54,76.4,88.9,89.4,57.6,48.6],
    "science": [85.10, 77.80, 91.54,91.54,88.9,89.4,57.6,48.6],
    "social": [85.10, 77.80, 91.54,76.4,88.9,89.4,57.6,48.6]}

df = pd.DataFrame(student_dict)
df.head(10)
```

```python
#duplicated method return true if duplicate placed otherwise false
print(df.duplicated())
print("")
```

```python
#drop duplicates remove the dupliacted rows
print(df.drop_duplicates())
```

```python
df.head(10)
```

```python
#using inplace true makes the raw dataframe effected
print(df.drop_duplicates(inplace=True))
```

```python
#syntax is important
print('and condition')
print(df.loc[(df['roll_no']>100) & (df["name"]=='dinesh')])
print("or condition ")
print(df.loc[(df['roll_no']>100) | (df["name"]=='dinesh')])
print('multiple and conditions')
print(df.loc[(df['roll_no']>100) & (df["name"]=='dinesh') & (df["telugu"]>85.5)])

#and so on
```
